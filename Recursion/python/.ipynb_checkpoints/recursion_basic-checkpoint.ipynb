```python
# tail recursion 
def fun(n):
    if n >0:
        print(n,end=" ")
        fun(n-1)
fun(3)
```

```python
# head recursion 
def fun(n):
    if n > 0:
        fun(n-1)
        print(n,end=" ")

fun(3)
```

```python

def fun(n):
    if n > 0:
        return fun(n-1)+ n
    return 0

print(fun(5))
```

```python
x = 0;
def fun(n):
    global x
    if n > 0:
        x+=1
        return fun(n-1)+ x
    return 0

print(fun(5))
```

```python
def treeRec(n):
    if n > 0:
        print(n)
        treeRec(n-1)
        treeRec(n-1)
treeRec(3)
```

```python
def indirectFunA(n):
    if  n > 0:
        print(n)
        indirectFunB(n-1)
def indirectFunB(n):
    if n > 1:
        print(n)
        indirectFunA(n//2)
indirectFunA(20)
```

```python
def nestedRec(n):
    if n > 100:
        return n - 10
    return nestedRec(nestedRec(n+11))
nestedRec(95)
```

```python
def sumOfN(n):
    return n * (n+1)//2

sumOfN(10)
```

```python
def sumOfNLoop(n):
    total = 0
    for i in range(n+1):
        total += i
    return total
    
sumOfNLoop(10)    
```

```python
def sumOfNRec(n):
    if (n == 0):
        return 0
    return sumOfNRec(n-1) + n
sumOfNRec(10)
```

```python
def factLoop(n):
    result = 1
    while n > 0:
        result *= n
        n -= 1
    return result
factLoop(5)
```

```python
def factRec(n):
    if n == 0:
        return 1
    return n* factRec(n-1)
factRec(5)
```

```python
def powerRec(m,n):
    if n == 0:
        return 1
    return m *powerRec(m,n-1)
powerRec(2,10)
```

```python
def power(m,n):
    if n == 0:
        return 1
    if n % 2 == 0:
        return power(m,n//2)
    return m*power(m*m,n-1//2)
        
```
