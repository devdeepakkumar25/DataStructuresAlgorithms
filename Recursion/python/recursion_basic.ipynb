```python
# tail recursion 
def fun(n):
    if n >0:
        print(n,end=" ")
        fun(n-1)
fun(3)
```

```python
# head recursion 
def fun(n):
    if n > 0:
        fun(n-1)
        print(n,end=" ")

fun(3)
```

```python
def nestedRec(n):
    if n > 100:
        return n - 10
    return nestedRec(nestedRec(n+11))
nestedRec(95)
```

```python
def sumOfN(n):
    return n * (n+1)//2

sumOfN(10)
```

```python
def sumOfNLoop(n):
    total = 0
    for i in range(n+1):
        total += i
    return total
    
sumOfNLoop(10)    
```

```python
def sumOfNRec(n):
    if (n == 0):
        return 0
    return sumOfNRec(n-1) + n
sumOfNRec(10)
```

```python
def factLoop(n):
    result = 1
    while n > 0:
        result *= n
        n -= 1
    return result
factLoop(5)
```

```python
def factRec(n):
    if n == 0:
        return 1
    return n* factRec(n-1)
factRec(5)
```

```python
def powerRec(m,n):
    if n == 0:
        return 1
    return m *powerRec(m,n-1)
powerRec(2,10)
```

```python
def powerFun(m,n):
    if n == 0:
        return 1
    if n % 2 == 0:
        return powerFun(m*m,n//2)
    return m*powerFun(m*m,(n-1)//2)

powerFun(2,9)
```

```python
def taylor(x,n):
    result = 1
    for i in range(1,n+1):
        fact = factRec(i)
        power = powerFun(x,i)
        result += power/fact

    return result
taylor(1,10)
```

```python
p = 1
f = 1

def taylorRec(x,n):
    global p
    global f 

    if n == 0:
        return 1
    else:
        r = taylorRec(x,n-1)
        p = p * x
        f = f * n 
    return r + p /f 
taylorRec(3,10)
        
        
    
```

```python
def taylorHornerRule(x,n):
    s = 1
    for i in range(n,0,-1):
        s = 1 + (x/i)*s 
    return s 
taylorHornerRule(1,10)
    
```

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-2) + fib(n-1)
fib(5)
```

```python
def fibItr(n):
    a = 0
    b = 1
    if n <=1:
        return n 
    for i in range(2,n+1):
        temp = a + b
        a = b 
        b = temp 
    return b
fibItr(7)
```

```python
def funRec(n):
    dp = [-1]*(n+1)
    if n <= 1:
        dp[n] = n
        return n 
    else:
        if dp[n-2] == -1:
            dp[n-2] = funRec(n-2)
        if dp[n-1] == -1:
            dp[n-1] = funRec(n-1)
        return dp[n-2] + dp[n-1]
        
funRec(7)
```

```python
def ncr(n,r):
    if (n == r or r == 0):
        return 1
    return ncr(n-1,r-1) + ncr(n-1,r)

ncr(3,2)
```
