```python
def linearSearch(arr,key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i 
    return -1

arr = [1,2,3,4,5,6]
print(linearSearch(arr,3))
```

def recLinearSearch(arr,key):
    def helper(arr,index,key):
        if index >= len(arr):
            return -1
        
        if arr[index] == key:
            return index
        return helper(arr,index+1,key)
    return helper(arr,0,key)

recLinearSearch(arr,1)

```python
def binarySearch(arr,key):
    left = 0
    right = len(arr) -1

    while left <= right:
        mid = (left + right)//2

        if arr[mid] == key:
            return mid
        if  key < arr[mid]:
            right = mid -1
        else:
            left = mid + 1
    return -1

arr = [x for x in range(10)]
arr.append(15)

binarySearch(arr,11)
```

```python
def recBinarySearch(arr,key):
    def helper(arr,left,right,key):
        if left > right:
            return -1
        mid = (left + right)//2
        if arr[mid] == key:
            return mid
        if key < arr[mid]:
           return helper(arr,left,mid-1,key)
        return helper(arr,mid+1,right,key)
        
    return helper(arr,0,len(arr)-1,key)

    
recBinarySearch(arr,1110)        
```

```python
def reverseArray(arr):
    left = 0
    right = len(arr) -1

    while left <= right:
        arr[left],arr[right] = arr[right],arr[left]
        left += 1
        right -= 1
        

arr = [1,2,3,4]        
reverseArray(arr)
arr
```

```python
def leftShift(arr):
    temp= arr[0]
    left = 1
    right = len(arr) -1

    while left <= right:
        arr[left-1] = arr[left]
        left +=1 
    arr[right] = temp

        
arr = [1,2,3,4,5]    
leftShift(arr)
arr
```

```python
def rightShift(arr):
    n = len(arr)-1
    temp = arr[n]

    for i in range(n,0,-1):
        arr[i] = arr[i-1]
    arr[0] = temp

arr = [1,2,3,4,5,6]        
rightShift(arr)
arr
```

```python
def insertInSortedArr(arr,val):
    x = val
    i = len(arr) -2
    while(arr[i] > val):
        arr[i+1] = arr[i]
        i -= 1
    arr[i+1] = x

arr = [1,2,3,4,6,7,8,0]
insertInSortedArr(arr,5)
arr
```

```python
def isSorted(arr):
    left = 1
    right = len(arr) -1 
    while left< right:
        if arr[left-1] > arr[left]:
            return False
        left += 1
    return True 

arr = [1,2,3,9,4,5,6,7]
print(isSorted(arr))
```

```python
def arrNegToPos(arr):
    left = 0 
    right = len(arr) -1

    while left < right:
    
        while  arr[left] <  0:
            left += 1
        while arr[right] >= 0:
            right -= 1
            
        if left < right:

            arr[left],arr[right] = arr[right],arr[left]

arr = [-6,3,-8,10,5,-7,-9,12,-4,2]
arrNegToPos(arr)
arr
```

```python
def merge(arr1,arr2):
    i = 0
    j =  0
    m = len(arr1)
    n = len(arr2)
    arr = []
    
    while i < m and j < n:
        if arr1[i] < arr2[j]:
            arr.append(arr1[i])
            i += 1
        else:
            arr.append(arr2[j])
            j += 1
    for index in range(i ,m):
        arr.append(arr1[index])
    for index in range(j,n):
        arr.append(arr2[index])
    print(arr)
    return arr

arr1 = [3,8,16,20,25]
arr2 = [4,10,12,22,23]
merge(arr1,arr2)
```

```python
def missingElementUsingHashing(arr):
    ht = [0]*(max(arr) + 1)
    for i in range(len(arr)):
        ht[arr[i]] += 1
    for i in range(arr[0],len(ht)):
        if ht[i] == 0:
            print(f"Missing Elment is : {i}")
missingElementUsingHashing(arr)
```

```python
def countDuplicate(arr):
    i = 1
    n = len(arr) -1
    while i < n:
        if arr[i-1] == arr[i]:
            j = i + 1
            while j < n and arr[j] == arr[i]:
                j += 1
            print(f"{arr[i]} is appering {j-i+1} times")
            i = j 
        else:       
            i += 1

arr = [ 3,6,8,8,10,12,15,15,15,20]
countDuplicate(arr)
```

```python
def duplicateUsingHash(arr):
    ht = [0]*(max(arr)+1)

    for i in range(len(arr)):
        ht[arr[i]] += 1

    for i in range(len(ht)):
        if (ht[i] > 1):
            print(f" Duplicate Element is: {i}")

duplicateUsingHash(arr)
```

```python
def duplicateInUnSorted(arr):
    n = len(arr)
    for i in range(len(arr)):
        count = 1
        if arr[i] != -1:
            for j in range(i+1,n):
                if arr[i] == arr[j]:
                    count += 1
                    arr[j] = -1
        if count > 1:
            print(f"{arr[i]} appers {count} times")

arr = [8,3,6,4,6,5,6,8,2,7]
duplicateInUnSorted(arr)
```

```python
def pairSumEqualTo(arr,k):
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i] + arr[j] == k:
                print(arr[i],arr[j],k)

arr = [6,3,8,10,16,7,5,2,9,14]

pairSumEqualTo(arr,10)
```
