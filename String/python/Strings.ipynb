```python
def changingCase(st):
    result = []
    for i in range(len(st)):
     
        ascii_code = ord(st[i])
     
        change_ascii_code = ascii_code - 32 
        alpha = chr(change_ascii_code)
        result.append(alpha)
    return "".join(result)
        
st = "welcome"
changingCase(st)
```

```python
def toggleCase(st):
    result = []
    for i in range(len(st)):
        ascii_code = ord(st[i])

        if 65 <= ascii_code <= 90:
            char = chr(ascii_code + 32)
            result.append(char)
        elif 97 <= ascii_code <= 122:
            char = chr(ascii_code - 32)
            result.append(char)
        else:
            result.append(st[i])
    return "".join(result)

st = "WeLcOmeAbC Def GHI mno"

toggleCase(st)
```

```python
def countVowels(st):
    vcount = 0 
    ccount = 0
    vowels = set("aeiouAEIOU")
    for i in range(len(st)):
        ascii_code = ord(st[i])
        if st[i] in vowels:
            vcount += 1
        elif 65 <= ascii_code <= 90 or 97 <= ascii_code <= 122:
            ccount += 1

    return (vcount,ccount)

st ="aeiouANBC1234"    
countVowels(st)
```

```python
def countWords(st):
    word = 0
    i = 0
    n = len(st)
    while i < n:
        while i < n and st[i] == " ":
            i += 1
        if i < n and st[i] != " ":
            word += 1
        while i < n and st[i] !=" ":
            i += 1
    return word
st ="How are You"
countWords(st)
```

```python
def validateString(st):
    for i in range(len(st)):
        ascii_code = ord(st[i])
        if not (48 <= ascii_code <= 57 or 65 <= ascii_code <= 90 or  97 <= ascii_code <= 122):
            return False
    return True
st = "abc!@"
validateString(st)
```

```python
def reverseStr(st):
    return st[::-1]

st = "abc"
reverseStr(st)
```

```python
def isPalindrome(st):
    i = 0 
    n = len(st) -1

    while i < n:
        if st[i] != st[n]:
            return False
        i += 1
        n -= 1
    return True 
    
st = "madam"
isPalindrome(st)
```

```python
def isDuplicate(st):
    s = set(st)
    return len(s) != len(st)

st = "finding"
# st = "abc"
isDuplicate(st)
```
