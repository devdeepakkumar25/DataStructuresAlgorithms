```python
def changingCase(st):
    result = []
    for i in range(len(st)):
     
        ascii_code = ord(st[i])
     
        change_ascii_code = ascii_code - 32 
        alpha = chr(change_ascii_code)
        result.append(alpha)
    return "".join(result)
        
st = "welcome"
changingCase(st)
```

```python
def toggleCase(st):
    result = []
    for i in range(len(st)):
        ascii_code = ord(st[i])

        if 65 <= ascii_code <= 90:
            char = chr(ascii_code + 32)
            result.append(char)
        elif 90 <= ascii_code <= 122:
            char = chr(ascii_code - 32)
            result.append(char)
        else:
            result.append(st[i])
    return "".join(result)

st = "WeLcOmeAbC Def GHI mno"

toggleCase(st)
```
